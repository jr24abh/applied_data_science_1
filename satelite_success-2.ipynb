# Statistics and Trend


#importing the packages
import numpy as np
import matplotlib.pyplot as  plt
from matplotlib.widgets import Cursor
import pandas as pd 
import seaborn as sns


```python
#reading the csv file using pandas
space_df=pd.read_csv(r"mission_launches.csv",index_col=['Unnamed: 0.1','Unnamed: 0'])

#cleaning/removing the emptycell or null value or empty values in table
space_df.dropna(axis=0,inplace=True)
space_df.dropna(axis=1,inplace=True)

#cleaning/removing the wrong formatted data
space_df['Date']= pd.to_datetime(space_df['Date'],format="mixed",utc=True)


#cleaning or removing the duplicates
space_df.drop_duplicates(inplace=True)

space_df.describe()

space_df['Price']=space_df['Price'].replace(',','',regex=False)
space_df['Price']=pd.to_numeric(space_df['Price'],errors='coerce')
```

```python
#co-relations

x=np.array(space_df['Organisation'].value_counts().index)
y=np.array(space_df['Organisation'].value_counts())

space_df['Organisation'].value_counts()
```

# plotting bar graphs

```python
#setting up width and height for figure
fig,ax=plt.subplots(figsize=(30,8))
cmap=plt.get_cmap("tab20")
colors = cmap(np.linspace(0, 1, len(x)))
# plotting the graph
ax.bar(x,y,label=x,color=colors)
# Presenting the lables in a presentable format
plt.xlabel('Organizations')
plt.ylabel('Mission launched')
plt.title('Organization and their missions')
plt.legend()
plt.grid(True)

plt.show()
```
